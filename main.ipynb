librosa **테스트**

```python
import librosa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import deque
import os 
```

```python
path = '/content/drive/MyDrive/Paper/datatest/vox1'
folder_list = os.listdir(path)
print("list = {}".format(folder_list))
```

```python
fname = []
ylist = []
srlist = []
fsize = []
duration = []
for folder in folder_list:
  fpath = librosa.util.find_files(path+'/'+folder)
  print(fpath)
  fsize.append(os.path.getsize(fpath[0]))
  y, sr = librosa.load(fpath[0])
  duration.append(librosa.core.get_duration(y=y,sr=sr,filename=fpath[0]))
  fname.append(fpath[0].split('\\')[-1])
  ylist.append(y)
  srlist.append(sr)
```
